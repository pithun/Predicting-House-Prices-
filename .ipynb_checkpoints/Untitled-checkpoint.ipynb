```python
def my_mean(lis):
    # removes outliers and then get the mean
    new_lis = []
    perc75o = lis.quantile(q= 0.75)
    perc25o = lis.quantile(q= 0.25)
    whiskero = (perc75o + 1.5*(perc75o-perc25o))
    bottom_whiskero = (perc25o - 1.5*(perc75o-perc25o))
    # print(whiskero, bottom_whiskero)
    for a in lis:
        if a < bottom_whiskero or a > whiskero:
            pass
        else:
            new_lis.append(a)
    return np.sum(new_lis)/len(new_lis)
```

```python
def city(val):
    # we see that ahmedabad has the costliest houses, followed by bangalore, then delhi and chennai which have
# similar distribution, hyderabad and kolkata and mumbai seem to have the same price distribution and 
# we see pune with the smallest distribution
    if val == 'Ahmedabad':
        return 5
    elif val == 'Bangalore':
        return 4
    elif val == 'Delhi' or val == 'Chennai':
        return 3
    elif val == 'Kolkata' or val == 'Mumbai' or val == 'Hyderabad': 
        return 2
    else:
        return 1
    
def room(val):
    if val == 1:
        return 0
    elif val == 4:
        return 1
    elif val == 5:
        return 2
    elif val == 2:
        return 3
    elif val == 3:
        return 4
    
def bathroom(val):
    if val == 1:
        return 0
    elif val == 4:
        return 1
    elif val == 5:
        return 2
    elif val == 3:
        return 3
    elif val == 2:
        return 4
    
def furnished_or_not(val):
    if val== 'Unfurnished':
        return 0
    #elif val== 'Semi-Furnished':
     #   return 2
    else:
        return 1

def layout(val):
    if val == 'BHK':
        return 1
    else:
        return 0
    
def property_type(val):
    if val == 'Apartment':
        return 5
    elif val == 'Independent House' or val == 'Independent Floor':
        return 4
    elif val == 'Penthouse':
        return 3
    elif val == 'Villa':
        return 2
    else: 
        return 0
    
def old_new(val):
    if val >= 4:
        return 1
    else:
        return 0
    
def outlier(val):
    #month = data[data.Offset_fault == a]
    perc75 = np.percentile(prices.security_deposit, 75)
    perc25 = np.percentile(prices.security_deposit, 25)
    whisker = (perc75 + 0.5*(perc75-perc25))
    bottom_whisker = (perc25 - 1.5*(perc75-perc25))
    if val <= bottom_whisker:
        return 1
    elif val >= whisker:
        return 1
    else: 
        return 0
    
```

```python
drop_cols = ['room', 'city', 'bathroom', 'layout_type', 'furnish_type', 'property_type', 'floor'
            , 'Property_ID']
# 90 for 2,3, 14 for 4,5, -0.068 for 9,10
base = 150
base1 = 3
def generate_logs(data, start, stop):
    for a in range(start, stop+1):
        data['log'+str(a)] = [np.round(math.log(c, a), 0) for c in data['security_deposit']]
        data['security_deposit'+str(a)] =  data['security_deposit']/data['log'+str(a)]
    return data
    
```

```python
def process(data):
    data['room_new'] = data['room'].map(room)
    data['city_new'] = data['city'].map(city)
    data['bathroom_new'] = data['bathroom'].map(bathroom)
    data['layout_new'] = data['layout_type'].map(layout)
    data['furnished_or_not'] = data['furnish_type'].map(furnished_or_not)
    data['property_type_new'] = data['property_type'].map(property_type)

    #data['area'] = data['area'].apply(lambda x: np.log10(x + 1))
    data['neighbourhood'] = data['neighbourhood'].apply(lambda x: np.log10(x + 1))
    #data['old_new'] = data['room'].map(old_new)
    data['log_t'] = np.log10(data['security_deposit'])
    #data['log_def'] = [math.log(c, base)  for c in data['security_deposit']]
    #data['log_def1'] = [math.log(c, base1)  for c in data['security_deposit']]
    #data['box_cox'] = stats.boxcox(data['security_deposit'], lmbda= 0.5)
    #data['box_cox1'] = stats.boxcox(data['security_deposit'], lmbda= -0.102)
    #data['diff'] = data['dep_months'] - data['log_t']
    #data['diff1'] = data['dep_months'] - data['log_def']
    #data['diff1a'] = data['dep_months'] - data['log_def1']

    #data['diff2'] = data['dep_months'] - data['box_cox']
    #data['security_deposit_log'] = data['security_deposit']/data['log_t']
    #data['security_deposit_box'] = data['security_deposit']/data['box_cox']
    #data['security_deposit_box1'] = data['security_deposit']/data['box_cox1']
    #data['security_deposit_log'] = data['security_deposit']/data['log_def']
    #data['security_deposit_log1'] = data['security_deposit']/data['log_def1']

    #data['outlier_detect'] = data.security_deposit.map(outlier)
    dum_fl = pd.get_dummies(data.floor)
    #dum_ci = pd.get_dummies(data.city)
    data = pd.concat([data, dum_fl], axis= 1)
    data = data.drop(drop_cols, axis = 1)
    #data = generate_logs(data, base1, base)
    return data

def scale(data1, data2):
    # Initialize a scaler, then apply it to the features
    scaler = MinMaxScaler() # default=(0, 1)
    float_var_names = []
    for a in data1.columns:
        if data1[a].dtype == 'float64' or data1[a].dtype == 'int64':
            float_var_names.append(a)
    #float_var_names.remove('price')
    data1[float_var_names] = scaler.fit_transform(data1[float_var_names])
    data2[float_var_names] = scaler.transform(data2[float_var_names])
    return data1, data2
```

```python
import math
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import MeanSquaredLogarithmicError
```

```python
def scale_datasets(X_train, X_test):
    standard_scaler = StandardScaler()
    x_train_scaled = pd.DataFrame(
      standard_scaler.fit_transform(X_train),
      columns=X_train.columns
  )
    x_test_scaled = pd.DataFrame(
      standard_scaler.transform(X_test),
      columns = X_test.columns
  )
    return x_train_scaled, x_test_scaled

x_train_scaled, x_test_scaled = scale_datasets(X_train, X_test)
```

```python
hidden_units1 = 160
hidden_units2 = 480
hidden_units3 = 256
learning_rate = 0.01
# Creating model using the Sequential in tensorflow
def build_model_using_sequential():
    model = Sequential([
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units2, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units3, kernel_initializer='normal', activation='relu'),
    Dense(1, kernel_initializer='normal', activation='linear')
  ])
    return model
```

```python
msle = MeanSquaredLogarithmicError()
model = build_model_using_sequential()
model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=learning_rate), 
    metrics=[msle]
)
# train the model
history = model.fit(
    x_train_scaled.values, 
    y_train.values, 
    epochs=20, 
    batch_size=64,
    validation_split=0.2
)
```

```python
def plot_history(history, key):
    plt.plot(history.history[key])
    plt.plot(history.history['val_'+key])
    plt.xlabel("Epochs")
    plt.ylabel(key)
    plt.legend([key, 'val_'+key])
    plt.show()
# Plot the history
plot_history(history, 'mean_squared_logarithmic_error')
```
