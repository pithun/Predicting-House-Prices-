```python
def my_mean(lis):
    # removes outliers and then get the mean
    new_lis = []
    perc75o = lis.quantile(q= 0.75)
    perc25o = lis.quantile(q= 0.25)
    whiskero = (perc75o + 1.5*(perc75o-perc25o))
    bottom_whiskero = (perc25o - 1.5*(perc75o-perc25o))
    # print(whiskero, bottom_whiskero)
    for a in lis:
        if a < bottom_whiskero or a > whiskero:
            pass
        else:
            new_lis.append(a)
    return np.sum(new_lis)/len(new_lis)
```

### Integer encoding functions

```python
def city(val):
    # we see that ahmedabad has the costliest houses, followed by bangalore, then delhi and chennai which have
# similar distribution, hyderabad and kolkata and mumbai seem to have the same price distribution and 
# we see pune with the smallest distribution
    if val == 'Ahmedabad':
        return 5
    elif val == 'Bangalore':
        return 4
    elif val == 'Delhi' or val == 'Chennai':
        return 3
    elif val == 'Kolkata' or val == 'Mumbai' or val == 'Hyderabad': 
        return 2
    else:
        return 1
    
def furnished_or_not(val):
    if val== 'Unfurnished':
        return 0
    #elif val== 'Semi-Furnished':
     #   return 2
    else:
        return 1

def layout(val):
    if val == 'BHK':
        return 1
    else:
        return 0
    
def property_type(val):
    if val == 'Apartment':
        return 5
    elif val == 'Independent House' or val == 'Independent Floor':
        return 4
    elif val == 'Penthouse':
        return 3
    elif val == 'Villa':
        return 2
    else: 
        return 0

def outlier(val):
    #month = data[data.Offset_fault == a]
    perc75 = np.percentile(prices.security_deposit, 75)
    perc25 = np.percentile(prices.security_deposit, 25)
    whisker = (perc75 + 0.5*(perc75-perc25))
    bottom_whisker = (perc25 - 1.5*(perc75-perc25))
    if val <= bottom_whisker:
        return 1
    elif val >= whisker:
        return 1
    else: 
        return 0
    
```

```python
drop_cols = ['city', 'layout_type', 'furnish_type', 'property_type', 'Property_ID', 'security_deposit', 'locality', 'floor']

base = 150
base1 = 3
def generate_logs(data, start, stop):
    # trying to see what logs best models the security deposit feature
    for a in range(start, stop+1):
        data['log'+str(a)] = [np.round(math.log(c, a), 0) for c in data['security_deposit']]
        data['security_deposit'+str(a)] =  data['security_deposit']/data['log'+str(a)]
    return data
    
```

## Creating function to process data

```python
def process(data):
    
    # feature engineering
    data['city_new'] = data['city'].map(city)
    data['layout_new'] = data['layout_type'].map(layout)
    data['furnished_or_not'] = data['furnish_type'].map(furnished_or_not)
    data['property_type_new'] = data['property_type'].map(property_type)
    data['log_t'] = np.log10(data['security_deposit'])
    data['locality_new'] = CountEncoder(cols= ['locality']).fit_transform(data.locality)

    # one hot encoding
    data['floor'] = data['floor'].astype('str')
    dum_fl = pd.get_dummies(data.floor)
    data = pd.concat([data, dum_fl], axis= 1)
    
    # dropping irrelevant columns
    data = data.drop(drop_cols, axis = 1)
    
    return data

def scale(data1, data2):
    # Initialize a scaler, then apply it to the features
    scaler = MinMaxScaler() # default=(0, 1)
    float_var_names = []
    for a in data1.columns:
        if data1[a].dtype == 'float64' or data1[a].dtype == 'int64':
            float_var_names.append(a)
    #float_var_names.remove('price')
    data1[float_var_names] = scaler.fit_transform(data1[float_var_names])
    data2[float_var_names] = scaler.transform(data2[float_var_names])
    return data1, data2

def log_t(col):
    return np.log10(col+1)
```

## Modelling

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(train_data.copy().drop(['price', 'dep_months'], axis = 1), train_data.copy().price, test_size=0.3,
                                                        random_state=42)
```

```python
# using base model to test performance

regressors = {
    'LGBM': LGBMRegressor(random_state=1),
    'Linear_regression':LinearRegression(),
    'Ridge': Ridge(),
    'XGB': XGBRegressor(random_state = 66),
    'cat': CatBoostRegressor(random_state = 77)
}

def train_return(X_train, X_test, y_train, y_test):
    df_models = pd.DataFrame(columns=['model', 'run_time', 'rmse', 'rmse_cv'])

    for key in regressors:

        start_time = time.time()

        regressor = regressors[key]
        model = regressor.fit(X_train, y_train.map(log_t))
        y_pred = model.predict(X_test)
        y_pred = [(10**x)-1 for x in y_pred]

        scores = cross_val_score(model, 
                             X_train, 
                             y_train,
                             scoring="neg_mean_squared_error", 
                             cv=5, verbose = 3, n_jobs = -1)

        row = {'model': key,
           'run_time': format(round((time.time() - start_time)/60,2)),
           'rmse': round(np.sqrt(mean_squared_error(y_test, y_pred))),
           'rmse_cv': round(np.mean(np.sqrt(-scores)))
        }

        df_models = df_models.append(row, ignore_index=True)
    return df_models
```

## Testing a Neural Network

```python
def scale_datasets(X_train, X_test):
    standard_scaler = StandardScaler()
    x_train_scaled = pd.DataFrame(
      standard_scaler.fit_transform(X_train),
      columns=X_train.columns
  )
    x_test_scaled = pd.DataFrame(
      standard_scaler.transform(X_test),
      columns = X_test.columns
  )
    return x_train_scaled, x_test_scaled

x_train_scaled, x_test_scaled = scale_datasets(X_train_iforest, X_test1)
```

```python
hidden_units1 = 160
hidden_units2 = 480
hidden_units3 = 256
learning_rate = 0.01
# Creating model using the Sequential in tensorflow
def build_model_using_sequential():
    model = Sequential([
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.3),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units2, kernel_initializer='normal', activation='relu'),
    Dropout(0.1),
    Dense(hidden_units3, kernel_initializer='normal', activation='relu'),
    Dense(1, kernel_initializer='normal', activation='linear')
  ])
    return model
```

```python
msle = MeanSquaredLogarithmicError()
model = build_model_using_sequential()
model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=learning_rate), 
    metrics=[msle]
)
# train the model
history = model.fit(
    x_train_scaled.values, 
    y_train_iforest.values, 
    epochs=20, 
    batch_size=64,
    validation_split=0.2
)
```

```python
def plot_history(history, key):
    plt.plot(history.history[key])
    plt.plot(history.history['val_'+key])
    plt.xlabel("Epochs")
    plt.ylabel(key)
    plt.legend([key, 'val_'+key])
    plt.show()
    
# Plotting the history
plot_history(history, 'mean_squared_logarithmic_error')
```
